```python
import pandas as pd
```

```python
raw_titanic = pd.read_csv('./data/train.csv', index_col='PassengerId')
raw_titanic.head(10)
```

```python
raw_titanic.drop(['Name','Ticket','Fare','Cabin','Embarked'],axis=1,inplace=True)
raw_titanic.head(10)
```

```python


titanic_data = pd.get_dummies(raw_titanic, columns=['Sex'], dtype=int)
titanic_data.head(10)
```

```python

titanic_data['age_label'] = pd.cut(titanic_data.Age, bins=[0, 2, 17, 65, 99], labels=['Baby', 'Child', 'Adult', 'Elderly'])
titanic_data.head(10)
```

```python

titanic_data.drop(['Age'], axis=1, inplace=True)
titanic_data.head(10)
```

```python
titanic_data.fillna('Adult', inplace=True)
titanic_data.head(10)
```

```python
titanic_data = pd.get_dummies(titanic_data, columns=['age_label'], dtype=int)
titanic_data.head(10)
```

```python
X = titanic_data.iloc[:, 1:]
y = titanic_data.iloc[:, 0]
```

```python
X[:5]
```

```python
y[:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, shuffle=True)
```

```python
print(X_train.shape, X_test.shape)

```

```python
print(y_train.shape, y_test.shape)
```

```python
from tensorflow import keras
```

```python
model = keras.Sequential()
model.add(keras.Input(shape=(X_train.shape[1],)))
model.add(keras.layers.Dense(24, activation='relu'))
model.add(keras.layers.Dropout(0.5))
model.add(keras.layers.Dense(12, activation='relu'))
model.add(keras.layers.Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer=keras.optimizers.Adam(learning_rate=0.005), metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=50, batch_size=10)
```

```python
score = model.evaluate(X_test, y_test)
print('Test Accuracy: %.2f' % (score[1]))
```
